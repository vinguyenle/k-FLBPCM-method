```python
import os
import random

dataset_base_path = "/scratch/data/filtered_crossval_bccr/work/"
#dataset_base_path = '/scratch/data/bccr_dataset_can_rad/work/'


dataset_list_prefix ='dataset_list_' # we'll generate lists of files with this prefix

N_sets = 5; # Number of sets to generate -> will split current dataset in this many sets
SEED = 444 # Change this number to anything else if you want a new randomisation

random.seed(SEED)  # Use this to get repeateable randomisation

def list_images(dir_path, stage_id='',file_ext='',):
    import os
    files = []; # Default return empty list if directory does not exist
    
    if os.path.isdir(dir_path):
        files = [f for f in os.listdir(dir_path) if (stage_id in f and \
                                                     f.endswith(file_ext) and \
                                                     os.path.isfile(dir_path+'/'+f) )]
    return files

dirs = [f for f in os.listdir(dataset_base_path) if (os.path.isdir(dataset_base_path+'/'+f)) ]

print 'Found the following classes:', dirs

stages = ['S1','S2','S3','S4'];
files_per_stage=[]
data_sets=[]
for my_dir in dirs:
    dir_path = dataset_base_path + '/' + my_dir
    print ' ----------------------------- '
    # For each new class
    files_per_stage.append([])
    for stage in stages:
        files_in_dir = list_images(dir_path,stage_id=stage) # files for this stage in dir
        random.shuffle(files_in_dir)
        files_per_stage[-1].append(files_in_dir)
        if len(files_in_dir):
            print (' [%10s] [%s]:  %d files , first img name=%s' %(my_dir,stage,\
                                                                 len(files_in_dir),\
                                                                 files_in_dir[0]))

# Now that we have our table with each class and stage, then generate the sets
print '\n\n'
for i in range(N_sets):
    data_set = []    
    for my_class in range(len(dirs)):
        for my_stage in range(len(stages)):
            # try and distribute same number of images per stage in each set
            n_img_per_set = int(len(files_per_stage[my_class][my_stage])/N_sets)
            start_index = i *n_img_per_set            
            end_index   = start_index + n_img_per_set;

            images  = files_per_stage[my_class][my_stage][start_index:end_index]
            image_label = dirs[my_class]
            print "DEBUG:[%d][%d]: n_img=%d => n_per_set=%d, i: %d:%d => %d imgs" \
                %(my_class,my_stage,len(files_per_stage[my_class][my_stage]),\
                  n_img_per_set, start_index, end_index, len(images))            
            # we keep label and image
            for image_name in images:
                data_set.append(image_label + ' ' + image_name)

    if len(data_set):
        # Now randomize
        random.shuffle(data_set)
        set_filename = dataset_base_path +"/" +dataset_list_prefix + str(i) + ".txt"
        print 'Writing Set[%d]: %s with %d images' %(i, set_filename, len(data_set))
        with open(set_filename,'w+') as f:
            for my_img in data_set:
                f.write(my_img + '\n')
                    
print 'Done '
```
